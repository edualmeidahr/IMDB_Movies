```python
import sys
from pathlib import Path
from collections import Counter

ROOT = Path('..').resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.utils import split_pipe_list
```

```python
import numpy as np

df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year

df['log_budget'] = np.log(df['budget'])
df['log_revenue'] = np.log(df['revenue'])
df['log_popularity'] = np.log1p(df['popularity'])
```

```python
import re

GENRE_TOP_K = 15
COMPANY_TOP_K = 20


def _ohe_col(prefix: str, label: str) -> str:
    safe = re.sub(r'[^\w]+', '_', str(label).strip()).strip('_')
    return f'{prefix}_{safe}'


# One-hot dos K gêneros mais frequentes
all_genres = []
for g in df['genres']:
    all_genres.extend(split_pipe_list(g))
top_genres = [g for g, _ in Counter(all_genres).most_common(GENRE_TOP_K)]

for g in top_genres:
    col = _ohe_col('genre', g)
    df[col] = df['genres'].apply(
        lambda x, gg=g: int(gg in split_pipe_list(x)) if pd.notna(x) else 0
    )

# One-hot das K produtoras mais frequentes + coluna "outras"
all_companies = []
for c in df['production_companies']:
    all_companies.extend(split_pipe_list(c))
top_companies = [c for c, _ in Counter(all_companies).most_common(COMPANY_TOP_K)]
top_companies_set = set(top_companies)

for co in top_companies:
    col = _ohe_col('company', co)
    df[col] = df['production_companies'].apply(
        lambda x, cc=co: int(cc in split_pipe_list(x)) if pd.notna(x) else 0
    )


def _has_other_company(cell, top_set):
    labs = split_pipe_list(cell)
    if not labs:
        return 0
    return int(any(lab not in top_set for lab in labs))


df['company_other'] = df['production_companies'].apply(
    lambda x: _has_other_company(x, top_companies_set) if pd.notna(x) else 0
)
```
